### Web scraping tables using pandas with function `p.read_html()`

```python
import pandas as pd
```

#### URL from Wikipedia with tables listing the largest banks

```python
URL = 'https://en.wikipedia.org/wiki/List_of_largest_banks'
```

```python
tables = pd.read_html(URL)
```

#### Get the first table from this site

```python
df = tables[0]
df
```

#### URL from Wikpedia with tables listing countries by GDP (nominal)

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
```

```python
all_tables = pd.read_html(url)
```

#### Get the table GDP (million US$) by country

```python
countries_by_gdp = all_tables[2]
countries_by_gdp
```
